```python
import numpy as np
import matplotlib.pyplot as plt

def display_npy_file(file_path):
    # Chargez le fichier `.npy`
    data = np.load(file_path)
    
    # Afficher quelques informations sur le fichier
    print(f"Shape des donn√©es: {data.shape}")
    print(data)
    
    # Pour visualiser correctement, s√©lectionnez une image dans le batch et canal
    image_to_display = data[0, 11]  # S√©lection du premier batch et canal

    plt.imshow(image_to_display, cmap='gray')
    plt.title(f"Visualisation de {file_path}")
    plt.colorbar()  # Ajoute une barre de couleur pour comprendre la plage des valeurs
    plt.show()

# Exemple d'utilisation
display_npy_file("preds/1/pano_semantic.npy")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def display_semantics(file_path):
    # Chargez le fichier `.npy`
    data = np.load(file_path)
    
    # Afficher quelques informations sur le fichier
    print(f"Shape des donn√©es: {data.shape}")
    print(f"Contenu des donn√©es (valeurs uniques): {np.unique(data)}")  # Peut aider √† comprendre la distribution des scores
    
    # V√©rification des dimensions
    num_objects, num_classes = data.shape
    
    # Afficher les scores pour un ou plusieurs objets
    for obj_index in range(min(5, num_objects)):  # Limit√© √† 5 pour √©viter trop d'affichages
        plt.bar(np.arange(num_classes), data[obj_index])
        plt.xlabel("Classe")
        plt.ylabel("Score")
        plt.title(f"Scores des classes pour l'objet {obj_index}")
        plt.show()

# Exemple d'utilisation
display_semantics("preds/1/semantic.npy")
```
